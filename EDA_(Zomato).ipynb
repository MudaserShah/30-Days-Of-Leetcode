```python
def rate_category(rating):
    if rating < 3.0:
        return 'Low'
    elif rating < 4.0:
        return 'Medium'
    else:
        return 'High'

Zomato['rating category'] = Zomato['rate (out of 5)'].apply(rate_category)
Zomato.groupby('rating category')['num of ratings'].mean()
```

```python
def popularity_level(num):
    if num < 100:
        return 'Low Popularity'
    elif num < 1000:
        return 'Medium Popularity'
    else:
        return 'Highly Rated'

Zomato['popularity'] = Zomato['num of ratings'].apply(popularity_level)
Zomato.groupby('popularity')['rate (out of 5)'].mean()
```
